```python
import pandas as pd

df = {}
csv_set = ["circuits", "constructor_results", "constructor_standings", "constructors", "driver_standings", "drivers", "lap_times", "pit_stops", "qualifying", "races", "results"]

for csv in csv_set:    
    path = f"/Users/pranav/Downloads/archive (2)/{csv}.csv"
    df[csv] = pd.read_csv(path)

# Selecting specific columns from each DataFrame
constructors_selected_columns = ['constructorId', 'constructorRef', 'name', 'nationality']
constructor_standings_selected_columns = ['driverStandingsId', 'raceId', 'driverId', 'points', 'position']
drivers_selected_columns = ['driverId', 'number', 'code', 'forename', 'surname', 'dob', 'nationality']
lap_times_selected_columns = ['raceId', 'driverId', 'lap', 'position', 'time']
pit_stops_selected_columns = ['raceId', 'driverId', 'stop', 'lap', 'time', 'duration', 'milliseconds']
qualifying_selected_columns = ['qualifyId', 'raceId', 'driverId', 'position', 'q3']
races_selected_columns = ['raceId', 'year', 'round', 'circuitId', 'date', 'time']
results_selected_columns = ['resultId', 'raceId', 'driverId', 'constructorId', 'grid', 'position', 'points', 'laps', 'fastestLapTime', 'fastestLapSpeed']
selected_columns_constructor_standings = ['constructorStandingsId', 'raceId', 'constructorId', 'position']
selected_columns_constructor_results = ['constructorResultsId', 'raceId', 'constructorId', 'points']
selected_columns = ['circuitId', 'circuitRef', 'name', 'location', 'country', 'alt']


# Create new DataFrames with selected columns
constructors_df = df['constructors'][constructors_selected_columns]
constructor_df = df['driver_standings'][constructor_standings_selected_columns]
drivers_df = df['drivers'][drivers_selected_columns]
lap_times_df = df['lap_times'][lap_times_selected_columns]
pit_stops_df = df['pit_stops'][pit_stops_selected_columns]
qualifying_df = df['qualifying'][qualifying_selected_columns]
races_df = df['races'][races_selected_columns]
results_df = df['results'][results_selected_columns]
constructor_standings_df = df['constructor_standings'][selected_columns_constructor_standings]
constructor_results_df = df['constructor_results'][selected_columns_constructor_results]
circuits_df = df['circuits'][selected_columns]


```

```python
import unicodedata

def remove_special_characters(text):
    normalized_text = unicodedata.normalize('NFKD', text)
    return normalized_text.encode('ascii', 'ignore').decode('utf-8')
```

```python
drivers_df['forename'] = drivers_df['forename'].apply(remove_special_characters)
drivers_df['surname'] = drivers_df['surname'].apply(remove_special_characters)
drivers_df['nationality'] = drivers_df['nationality'].apply(remove_special_characters)
drivers_df.head(10)
```

```python

def replace_na(cell_value):
    if '\\N' in str(cell_value):
        return None
    else:
        return cell_value
```

```python
column_to_process = 'number'
drivers_df[column_to_process] = drivers_df[column_to_process].apply(replace_na)

column_to_process = 'code'
drivers_df[column_to_process] = drivers_df[column_to_process].apply(replace_na)

print(drivers_df)
```

```python
circuits_df['circuitRef'] = circuits_df['circuitRef'].apply(remove_special_characters)
circuits_df['name'] = circuits_df['name'].apply(remove_special_characters)
circuits_df['country'] = circuits_df['country'].apply(remove_special_characters)
print(circuits_df.head(10))
```

```python
column_to_process = 'alt'

circuits_df[column_to_process] = circuits_df[column_to_process].apply(replace_na)
print(circuits_df)
```

```python
constructors_df['constructorRef'] = constructors_df['constructorRef'].apply(remove_special_characters)
constructors_df['name'] = constructors_df['name'].apply(remove_special_characters)
constructors_df['nationality'] = constructors_df['nationality'].apply(remove_special_characters)
print(constructors_df.head(10))
```

```python
column_to_process = 'time'
races_df[column_to_process] = races_df[column_to_process].apply(replace_na)

print(races_df)
```

```python
column_to_process = 'position'
results_df[column_to_process] = results_df[column_to_process].apply(replace_na)
column_to_process = 'fastestLapSpeed'
results_df[column_to_process] = results_df[column_to_process].apply(replace_na)
column_to_process = 'fastestLapTime'
results_df[column_to_process] = results_df[column_to_process].apply(replace_na)
column_to_process = 'points'
results_df[column_to_process] = results_df[column_to_process].apply(replace_na)
column_to_process = 'laps'
results_df[column_to_process] = results_df[column_to_process].apply(replace_na)
print(results_df)
```
